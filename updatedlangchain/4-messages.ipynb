Messages: messages are the fundamenta unit of context for models in Langchain. They represents the input and output of models, carrying both the content and metadata needed to represents the state of a converstaion when interacting with an LLM. Messages are objects that contains:

role     -> Identifies who sent the message (user, assistant, system, tool)

content  -> The actual text or data being communicated

metadata -> Additional information such as token usage,timestamps, tool calls, or execution details

```python
## model initialization
import os
from langchain.chat_models import init_chat_model

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

model = init_chat_model("google_genai:gemini-2.5-flash-lite")
response = model.invoke("why do parrot talk?")
response
```

Message Prompts

Instead of a single prompt, we can pass a list of messages to maintain conversation context.

Message Types

-System Message -> Defines the AI's behavior and instructions.
-Human Message -> Input from the user.
-AI Message -> Response generated by the model.
-Tool Message -> Output returned by a tool.

System Message
Provides instructions, rules, and context that guide the model's behavior throughout the conversation.

Human Message
Represents a message or question sent by the user to the AI.

AI Message
Represents the response generated by the AI model.

Tool Message
Contains the result returned by an external tool that the AI has called.

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage(content="You are a helpful AI assistant."),
    HumanMessage(content="What is LangGraph?"),
    AIMessage(content="LangGraph is a framework for building stateful AI agents.")
]

response = model.invoke(messages)
response.content
```
