Structured Output

Structured output allows the model to return responses in a predefined format, such as JSON or a schema, making the output easier to validate, parse, and use in applications.

Benefits:
- Consistent response format
- Easy data extraction
- Better integration with applications
- Reduces parsing errors

Pydantic

Pydantic is a Python library used to define, validate, and structure data using Python classes and type hints.

Benefits:
- Data validation
- Type checking
- Structured outputs
- Easy JSON conversion

```python
## model initialization
import os
from langchain.chat_models import init_chat_model

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

model = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
from pydantic import BaseModel,Field

class Movie(BaseModel):
    title:str=Field(description="The title of the movie")
    year:int=Field(description="This year the movie was released")
    director:str=Field(description="The director of the movie")
    rating:str=Field(description="The movie rating out of 10")
```

```python
model_with_structure = model.with_structured_output(Movie)
```

```python
model.invoke("Provide details about the movie Inception")
```

```python
response = model_with_structure.invoke("Provide details about the movie Inception")
response
```

## Message output alongside parsed structure

```python
from pydantic import BaseModel,Field

class Movie(BaseModel):
    title:str=Field(description="The title of the movie")
    year:int=Field(description="This year the movie was released")
    director:str=Field(description="The director of the movie")
    rating:str=Field(description="The movie rating out of 10")

model_with_structure = model.with_structured_output(Movie, include_raw=True)

response = model_with_structure.invoke("Provide details about the movie Inception")
response
```

Nested Structure

```python
class Actor(BaseModel):
    name:str
    role:str

class MovieDetails(BaseModel):
    title:str
    year:int
    cast:list[Actor] #list of actors
    genres:list[str]
    budget:float | None = Field(None , description="Budget in millions USD")

model_with_structure = model.model_with_structure_output(MovieDetails)

response = model_with_structure.invoke("Provide details about the movie Inception")
response
```

TypeDict

TypeDict provides a simpler alternative using Pythons built in typing, ideal when you dont need runtime validation

```python
from typing_extensions import TypedDict,Annotated

class MovieDict(TypedDict):
    """A movie with details."""
    title:Annotated[str, ..., "The title of the movie"]
    year:Annotated[int, ..., "The year the movie was released"]
    director:Annotated[str, ..., "The director of the movie"]
    rating:Annotated[float, ..., "The movie rating out of 10"]

model_withtypedict = model.with_structured_output(MovieDict)
response = model_withtypedict("Please provide the details of the movie avengers")
response
```
