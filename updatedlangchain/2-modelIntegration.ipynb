Streamin and Batch

Streaming

n Agentic AI, streaming is the process of sending intermediate outputs, tool results, reasoning steps, or generated tokens to the user in real time while the agent is executing a task, instead of waiting for the entire workflow to finish before returning a response.

```python
## in this case you have to wait til the response
model.invoke("wite me a 200 words paragraph on artifical intellgence")
```

```python
## in this case we dont have to wait til the response
for chunk in model.stream("wite me a 200 words paragraph on artifical intellgence"):
    print(chunk.text, end="|",flush=True)
```

Batch
batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel:

```python
responses = model.batch([
    "why do parrots have colorful features?",
    "How do airplane fly?",
    "what is quantum computing?"
])
for response in responses:
    print(response)
```
